## Static Variable

```python
csv_file = "paper_top_5000_apps.csv"
source_dir = r"C:\Users\ASUS\anaconda3\phd_implement\photography_category"
dest_dir = r"C:\Users\ASUS\anaconda3\metaLeak-ml-apk-upload-client\upload_apk_no_leak"
no_leak_apk_file = "no_leak_apk.csv"
```

## Function

```python
# def get_apkfile_names(csv_file):
#     # Read CSV file into DataFrame
#     df = pd.read_csv(csv_file)

#     # Filter rows where metadata_leak is "No Leak Sensitive Metadata" and get values of apkfile_name
#     apkfile_names = df.loc[df['metadata_leak'] == 'No Leak Sensitive Metadata', 'apkfile_name'].tolist()

#     return apkfile_names
def get_apkfile_names(csv_file, output_csv):
    # Read CSV file into DataFrame
    df = pd.read_csv(csv_file)

    # Filter rows where metadata_leak is "No Leak Sensitive Metadata"
    no_leak_df = df[df['metadata_leak'] == 'No Leak Sensitive Metadata']

    # Get values of apkfile_name
    apkfile_names = no_leak_df['apkfile_name'].tolist()

    # Write filtered rows to a new CSV file
    no_leak_df.to_csv(output_csv, index=False)

    return apkfile_names
def copy_apk_file(apk_file, source_dir, dest_dir):

    # Check if the APK file exists in the source directory
    src_path = os.path.join(source_dir, apk_file)
    dest_path = os.path.join(dest_dir, apk_file)

    if os.path.exists(src_path):
        # Check if the file exists in the destination directory
        if not os.path.exists(dest_path):
            # Copy the APK file to the destination directory
            shutil.copyfile(src_path, dest_path)
            print(f"Copied {apk_file} to {dest_dir}")
        else:
            print(f"{apk_file} already exists in {dest_dir}")
    else:
        print(f"{apk_file} does not exist in {source_dir}")
```

## Main

```python
no_leak_arr = get_apkfile_names(csv_file,no_leak_apk_file)
print(len(no_leak_arr))
for i in range(len(no_leak_arr)):
    print("-------------------------------Loop "+str(i)+"-------------------------------")
    apk_filename = no_leak_arr[i]
    print("APK file name: "+apk_filename)
    copy_apk_file(apk_filename,source_dir,dest_dir)
    #break
```
