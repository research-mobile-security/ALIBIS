```python
import json
import os
```

```python
def rewrite_json_file(input_file_path, output_file_path):
    # Open the input JSON file
    with open(input_file_path, 'r') as input_file:
        data = json.load(input_file)

    # Extract the dictionary from the list
    data_dict = data[0]

    # Write the dictionary to the output file
    with open(output_file_path, 'w') as output_file:
        json.dump(data_dict, output_file, indent=4)  
# Function list all file in directory to list
def list_files_in_directory(directory_path):
    file_list = os.listdir(directory_path)
    file_list = [file for file in file_list if os.path.isfile(os.path.join(directory_path, file))]
    return file_list
```

```python
director_path = r"C:\Users\ASUS\anaconda3\metaLeak-ml-llm-rag\exif_keyword_code_block"
rewrite_path = r"C:\Users\ASUS\anaconda3\metaLeak-ml-llm-rag\exif_keyword_code_block_re"
```

```python
file_array = list_files_in_directory(director_path)
#print(file_array)
#print(len(file_array))
for i in range(len(file_array)):
    print("-----------------------------------Loop-"+str(i)+"-----------------------------------")
    file_path_read = director_path+"\\"+file_array[i]
    file_path_write = rewrite_path+"\\"+file_array[i]
    print("File path read: "+file_path_read)
    print("File path write: "+file_path_write)
    rewrite_json_file(file_path_read,file_path_write)
```
