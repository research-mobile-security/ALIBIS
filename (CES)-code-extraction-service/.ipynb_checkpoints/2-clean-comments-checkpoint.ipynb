# CLEAN COMMENTS

```python
import os
import re
```

## 1. FUNCTION

```python
# Function get directory
def get_all_directories(path):
    directories = []
    for entry in os.listdir(path):
        full_path = os.path.join(path, entry)
        if os.path.isdir(full_path):
            directories.append(entry)
    return directories
# Function get all file in directory
def get_all_files_in_directory(directory_path):
    # Initialize an empty list to store file names
    file_list = []

    # Iterate through all files and directories in the specified directory
    for item in os.listdir(directory_path):
        # Construct the full path of the item
        full_path = os.path.join(directory_path, item)

        # Check if the item is a file (not a directory)
        if os.path.isfile(full_path):
            # Add the file to the list
            file_list.append(full_path)
    return file_list

def remove_java_comments(input_file_path):
    try:
        with open(input_file_path, 'r') as input_file:
            lines = input_file.readlines()

        with open(input_file_path, 'w') as output_file:
            for line in lines:
                if not line.strip().startswith('//'):
                    output_file.write(line)

        print(f"Comments removed successfully from '{input_file_path}'.")
    except Exception as e:
        print(f"An error occurred: {e}")
```

## 2. STATIC VARIABLES

```python
exif_java_directory = r".\exif-java-no-leak"
```

```python
directories = get_all_directories(exif_java_directory)
for i in range(len(directories)):
    print("----------------------------------------"+str(i)+"----------------------------------------")
    directory = exif_java_directory+"\\"+str(directories[i])
    arr_java_file = get_all_files_in_directory(directory)
    print(arr_java_file)
    for j in range(len(arr_java_file)):
        java_file = arr_java_file[j]
        remove_java_comments(java_file)
```
