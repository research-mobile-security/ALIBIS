# CODE SEARCH NO LEAK

```python
import os
import shutil
import pandas as pd
```

## 1. FUNCTION

```python
# Function search file path that contain keywork
def find_files_with_keyword(directory, keyword):
    matching_files = []
    for root, _, files in os.walk(directory):
        for file in files:
            file_path = os.path.join(root, file)
            try:
                with open(file_path, 'r', encoding='utf-8') as f:
                    content = f.read()
                    if keyword in content:
                        matching_files.append(file_path)
            except Exception as e:
                print(f"Error reading {file_path}: {e}")
    return matching_files
# Function get directory
def get_all_directories(path):
    directories = []
    for entry in os.listdir(path):
        full_path = os.path.join(path, entry)
        if os.path.isdir(full_path):
            directories.append(entry)
    return directories
# Function create directory
def create_directory(path):
    if not os.path.exists(path):
        os.makedirs(path)
        print("Create directory "+path+" success!")
    else:
        print("Directory "+path+" exist!")
# Copy file
def copy_file(source_path, destination_path):
    try:
        # Copy the file from source to destination
        shutil.copy(source_path, destination_path)
        print(f"File '{source_path}' copied successfully to '{destination_path}'.")
    except FileNotFoundError:
        print("Source file does not exist.")
    except PermissionError:
        print("Permission denied to copy the file.")
    except shutil.SameFileError:
        print("Source and destination files are the same.")
    except Exception as e:
        print(f"An error occurred while copying the file: {e}")
# Update CSV
def update_csv_exif_file_no(csv_file, apkfile_name, exif_file_no):
    try:
        # Read the CSV file into a DataFrame
        df = pd.read_csv(csv_file)
        
        # Update the "exif_file_no" column where "apkfile_name" matches the given value
        df.loc[df['apkfile_name'] == apkfile_name, 'exif_file_no'] = int(exif_file_no)
        
        # Write the updated DataFrame back to the CSV file
        df.to_csv(csv_file, index=False)
        
        print(f"CSV file '{csv_file}' updated successfully.")
    except FileNotFoundError:
        print("CSV file not found.")
    except Exception as e:
        print(f"An error occurred while updating the CSV file: {e}")
```

## 2. STATIC VARIABLE

```python
decompile_source_code = r".\decompile-code-no-leak"
keyword = "exif"
exif_java_directory = r".\exif-java-no-leak"
csv_file = "no_leak_apk_low_size.csv"
```

## 3. MAIN

```python
# 1. Get all directory name
arr_java_directory = get_all_directories(decompile_source_code)
# 2. Search file .java related to "EXIF"
for i in range(len(arr_java_directory)):
    print("---------------------------------Loop "+str(i)+": "+str(arr_java_directory[i])+"---------------------------------")
    directory_name = (str(arr_java_directory[i])).split('-java')[0]
    directory_name = exif_java_directory+"\\"+directory_name
    create_directory(directory_name)
    source_path = decompile_source_code+"\\"+str(arr_java_directory[i])
    apkfile_name = (str(arr_java_directory[i])).split('-java')[0]+".apk"
    print(apkfile_name)
    exif_files = find_files_with_keyword(source_path, keyword)
    exif_file_no = str(len(exif_files))
    print("Number of exif-related file: "+exif_file_no)
    update_csv_exif_file_no(csv_file, apkfile_name, exif_file_no)
    if exif_files:
        print("Files containing the keyword 'Exif':")
        for file_path in exif_files:
            print(file_path)
            copy_file(file_path,directory_name)
    else:
        print("No files found containing the keyword 'Exif'.")
```
