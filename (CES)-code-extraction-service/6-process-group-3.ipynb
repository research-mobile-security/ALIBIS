```python
import re
import json
import os
```

## 1. FUNCTION

```python
def get_all_directories(path):
    directories = []
    for entry in os.listdir(path):
        full_path = os.path.join(path, entry)
        if os.path.isdir(full_path):
            directories.append(entry)
    return directories
# Get all file in directory
def get_all_files_in_directory(directory_path):
    # Initialize an empty list to store file names
    file_list = []

    # Iterate through all files and directories in the specified directory
    for item in os.listdir(directory_path):
        # Construct the full path of the item
        full_path = os.path.join(directory_path, item)

        # Check if the item is a file (not a directory)
        if os.path.isfile(full_path):
            # Add the file to the list
            file_list.append(full_path)
    return file_list
def extract_static_variables(file_path):
    static_variables = []
    within_static_block = False

    with open(file_path, 'r') as file:
        for line in file:
            # Check if the line starts the static block
            if re.match(r'\s*static\s*\{', line):
                within_static_block = True
                continue

            # Check if the line ends the static block
            if within_static_block and re.match(r'\s*\}\s*', line):
                within_static_block = False
                break

            # If we're within the static block, add the line to the list
            if within_static_block:
                static_variables.append(line.strip())

    return static_variables
# Function search parameter between ""
def extract_values_with_quotes(arr):
    values = []
    for item in arr:
        matches = re.findall(r'"([^"]*)"', item)
        if matches:
            values.extend(matches)
    return values
# Function append dict to dict
def append_dict_to_dict(dict1, dict2):
    for key, value in dict2.items():
        if key in dict1:
            # If the key exists in dict1, append the value to its list
            dict1[key].append(value)
        else:
            # If the key doesn't exist in dict1, create a new list with the value
            dict1[key] = [value]
# Function remove duplicate value in list
def remove_dupicate_list(lst):
    unique_values = []
    seen = set()
    for item in lst:
        if item not in seen:
            unique_values.append(item)
            seen.add(item)
    return unique_values
# Function remove empty value
def remove_empty_strings(lst):
    return [item for item in lst if item != '']
# Append dict to JSON file
def append_dict_to_json_file(data_dict, file_path):
    try:
        # Read existing JSON data from file
        with open(file_path, 'r') as file:
            existing_data = json.load(file)
    except FileNotFoundError:
        # If file doesn't exist, initialize with an empty list
        existing_data = []

    # Append new dictionary to existing data
    existing_data.append(data_dict)

    # Write updated data back to JSON file
    with open(file_path, 'w') as file:
        json.dump(existing_data, file, indent=4)
def remove_file_if_exists(file_path):
    if os.path.exists(file_path):
        os.remove(file_path)
        print(f"File '{file_path}' removed successfully.")
    else:
        print(f"File '{file_path}' does not exist.")
# Function search line
def search_keyword_in_file(file_path, keyword):
    # List to store lines containing the keyword
    result = []

    # Open the file for reading
    with open(file_path, 'r') as file:
        # Iterate through each line in the file
        for line in file:
            # Check if the keyword (case insensitive) is in the line
            if keyword.lower() in line.lower():
                # If found, append the line to the result list
                line = line.strip()
                result.append(line.rstrip())  # Remove trailing newline character

    return result
```

## 2. STATIC VARIABLE

```python
group_3_path = r'.\grouping\group-3'
class_name_arr = [
    "ArrayList",
    "LinkedList",
    "HashMap",
    "HashSet",
    "StringBuilder"
]
method_name_arr = [
    "add(",
    "get(",
    "set(",
    "remove(",
    "clear(",
    "put(",
    "access(",
    "contains(",
    "append(",
    "delete(",
    "insert("
]
```

## 3. MAIN

```python
group_3_directories = get_all_directories(group_3_path)
for i in range(len(group_3_directories)):
    print("------------------------------Loop-"+str(i)+"------------------------------")
    code_dir = group_3_path+"\\"+str(group_3_directories[i])
    print(code_dir)
    json_file_path = code_dir+".json"
    
    remove_file_if_exists(json_file_path)
    arr_file = get_all_files_in_directory(code_dir)
    
    final_static_parameter = {}
    static_parameter_dict = {}
    
    class_name_ArrayList_dict = {}
    final_class_name_ArrayList_dict = {}
    
    class_name_LinkedList_dict = {}
    final_class_name_LinkedList_dict = {}
    
    class_name_HashMap_dict = {}
    final_class_name_HashMap_dict = {}
    
    class_name_HashSet_dict = {}
    final_class_name_HashSet_dict = {}
    
    class_name_StringBuilder_dict = {}
    final_class_name_StringBuilder_dict = {}
    
    method_name_add_dict = {}
    final_method_name_add_dict = {}
    
    method_name_get_dict = {}
    final_method_name_get_dict = {}
    
    method_name_set_dict = {}
    final_method_name_set_dict = {}
    
    method_name_remove_dict = {}
    final_method_name_remove_dict = {}

    method_name_clear_dict = {}
    final_method_name_clear_dict = {}
    
    method_name_put_dict = {}
    final_method_name_put_dict = {}
    
    method_name_access_dict = {}
    final_method_name_access_dict = {}

    method_name_contains_dict = {}
    final_method_name_contains_dict = {}

    method_name_append_dict = {}
    final_method_name_append_dict = {}

    method_name_delete_dict = {}
    final_method_name_delete_dict = {}
    
    method_name_insert_dict = {}
    final_method_name_insert_dict = {}

    for j in range(len(arr_file)):
        file_name = str(arr_file[j])
        print(file_name)
        static_block = extract_static_variables(file_name)
        static_parameter = extract_values_with_quotes(static_block)
        static_parameter = remove_dupicate_list(static_parameter)
        static_parameter = remove_empty_strings(static_parameter)
        if(len(static_parameter)>0):
            static_parameter_dict["static_parameter"] = static_parameter
            append_dict_to_dict(final_static_parameter,static_parameter_dict)
        # Process class name
        for x in range(len(class_name_arr)):
            class_name = class_name_arr[x]
            #print(class_name)
            class_name_line = search_keyword_in_file(file_name,class_name)
            # Get parameter in method name
            #class_name_line = extract_values_with_quotes(class_name_line)
            #print(class_name_line)
            # Remove duplicate
            class_name_line = remove_dupicate_list(class_name_line)
            #print(len(class_name_line))
            #print(class_name_line)
            if len(class_name_line) >0:
                if class_name=="ArrayList":
                    class_name_ArrayList_dict["ArrayList"] = class_name_line
                    append_dict_to_dict(final_class_name_ArrayList_dict,class_name_ArrayList_dict)
                elif class_name=="LinkedList":
                    class_name_LinkedList_dict["LinkedList"] = class_name_line
                    append_dict_to_dict(final_class_name_LinkedList_dict,class_name_LinkedList_dict)
                elif class_name=="HashMap":
                    class_name_HashMap_dict["HashMap"] = class_name_line
                    append_dict_to_dict(final_class_name_HashMap_dict,class_name_HashMap_dict)
                elif class_name=="HashSet":
                    class_name_HashSet_dict["HashSet"] = class_name_line
                    append_dict_to_dict(final_class_name_HashSet_dict,class_name_HashSet_dict)
                elif class_name=="StringBuilder":
                    class_name_StringBuilder_dict["StringBuilder"] = class_name_line
                    append_dict_to_dict(final_class_name_StringBuilder_dict,class_name_StringBuilder_dict)             
        # Process method name
        for y in range(len(method_name_arr)):
            method_name = method_name_arr[y]
            method_name_line = search_keyword_in_file(file_name,method_name)
            # Get parameter in method name
            #method_name_line = extract_values_with_quotes(method_name_line)
            # Remove duplicate
            method_name_line = remove_dupicate_list(method_name_line)
            if len(method_name_line) >0:
                if method_name=="add(":
                    method_name_add_dict["add_method"] = method_name_line
                    append_dict_to_dict(final_method_name_add_dict,method_name_add_dict)
                elif method_name=="get(":
                    method_name_get_dict["get_method"] = method_name_line
                    append_dict_to_dict(final_method_name_get_dict,method_name_get_dict)
                elif method_name=="set(":
                    method_name_set_dict["set_method"] = method_name_line
                    append_dict_to_dict(final_method_name_set_dict,method_name_set_dict)
                elif method_name=="remove(":
                    method_name_remove_dict["remove_method"] = method_name_line
                    append_dict_to_dict(final_method_name_remove_dict,method_name_remove_dict)
                elif method_name=="clear(":
                    method_name_clear_dict["clear_method"] = method_name_line
                    append_dict_to_dict(final_method_name_clear_dict,method_name_clear_dict)
                elif method_name=="put(":
                    method_name_put_dict["put_method"] = method_name_line
                    append_dict_to_dict(final_method_name_put_dict,method_name_put_dict)
                elif method_name=="access(":
                    method_name_access_dict["access_method"] = method_name_line
                    append_dict_to_dict(final_method_name_access_dict,method_name_access_dict)
                elif method_name=="contains(":
                    method_name_contains_dict["contains_method"] = method_name_line
                    append_dict_to_dict(final_method_name_contains_dict,method_name_contains_dict)
                elif method_name=="append(":
                    method_name_append_dict["append_method"] = method_name_line
                    append_dict_to_dict(final_method_name_append_dict,method_name_append_dict)
                elif method_name=="delete(":
                    method_name_delete_dict["delete_method"] = method_name_line
                    append_dict_to_dict(final_method_name_delete_dict,method_name_delete_dict)
                elif method_name=="insert(":
                    method_name_insert_dict["insert_method"] = method_name_line
                    append_dict_to_dict(final_method_name_insert_dict,method_name_insert_dict)
    print("~~~~~~~~~~~~~Static{}~~~~~~~~~~~~~")
    print(final_static_parameter)
    
    print("~~~~~~~~~~~~~ArrayList~~~~~~~~~~~~~")
    print(final_class_name_ArrayList_dict)
    print("~~~~~~~~~~~~~LinkedList~~~~~~~~~~~~~")
    print(final_class_name_LinkedList_dict)
    print("~~~~~~~~~~~~~HashMap~~~~~~~~~~~~~")
    print(final_class_name_HashMap_dict)
    print("~~~~~~~~~~~~~HashSet~~~~~~~~~~~~~")
    print(final_class_name_HashSet_dict)
    print("~~~~~~~~~~~~~StringBuilder~~~~~~~~~~~~~")
    print(final_class_name_StringBuilder_dict)
    
    print("~~~~~~~~~~~~~Add_method~~~~~~~~~~~~~")
    print(final_method_name_add_dict)
    print("~~~~~~~~~~~~~Get_method~~~~~~~~~~~~~")
    print(final_method_name_get_dict)
    print("~~~~~~~~~~~~~Set_method~~~~~~~~~~~~~")
    print(final_method_name_set_dict)
    print("~~~~~~~~~~~~~Remove_method~~~~~~~~~~~~~")
    print(final_method_name_remove_dict)
    print("~~~~~~~~~~~~~Clear_method~~~~~~~~~~~~~")
    print(final_method_name_clear_dict)
    print("~~~~~~~~~~~~~Put_method~~~~~~~~~~~~~")
    print(final_method_name_put_dict)
    print("~~~~~~~~~~~~~Access_method~~~~~~~~~~~~~")
    print(final_method_name_access_dict)
    print("~~~~~~~~~~~~~Contains_method~~~~~~~~~~~~~")
    print(final_method_name_contains_dict)
    print("~~~~~~~~~~~~~Append_method~~~~~~~~~~~~~")
    print(final_method_name_append_dict)
    print("~~~~~~~~~~~~~Append_method~~~~~~~~~~~~~")
    print(final_method_name_delete_dict)
    print("~~~~~~~~~~~~~Insert_method~~~~~~~~~~~~~")
    print(final_method_name_insert_dict)
    # static{}
    if len(final_static_parameter)>0:
        append_dict_to_json_file(final_static_parameter,json_file_path)
    # class    
    if len(final_class_name_ArrayList_dict)>0:
        append_dict_to_json_file(final_class_name_ArrayList_dict,json_file_path)
    if len(final_class_name_LinkedList_dict)>0:
        append_dict_to_json_file(final_class_name_LinkedList_dict,json_file_path)
    if len(final_class_name_HashMap_dict)>0:
        append_dict_to_json_file(final_class_name_HashMap_dict,json_file_path)
    if len(final_class_name_HashSet_dict)>0:
        append_dict_to_json_file(final_class_name_HashSet_dict,json_file_path)
    if len(final_class_name_StringBuilder_dict)>0:
        append_dict_to_json_file(final_class_name_StringBuilder_dict,json_file_path)
    # method
    if len(final_method_name_add_dict)>0:
        append_dict_to_json_file(final_method_name_add_dict,json_file_path)
    if len(final_method_name_get_dict)>0:
        append_dict_to_json_file(final_method_name_get_dict,json_file_path)
    if len(final_method_name_set_dict)>0:
        append_dict_to_json_file(final_method_name_set_dict,json_file_path)
    if len(final_method_name_remove_dict)>0:
        append_dict_to_json_file(final_method_name_remove_dict,json_file_path)
    if len(final_method_name_clear_dict)>0:
        append_dict_to_json_file(final_method_name_clear_dict,json_file_path)
    if len(final_method_name_put_dict)>0:
        append_dict_to_json_file(final_method_name_put_dict,json_file_path)
    if len(final_method_name_access_dict)>0:
        append_dict_to_json_file(final_method_name_access_dict,json_file_path)
    if len(final_method_name_contains_dict)>0:
        append_dict_to_json_file(final_method_name_contains_dict,json_file_path)
    if len(final_method_name_append_dict)>0:
        append_dict_to_json_file(final_method_name_append_dict,json_file_path)
    if len(final_method_name_delete_dict)>0:
        append_dict_to_json_file(final_method_name_delete_dict,json_file_path)
    if len(final_method_name_insert_dict)>0:
        append_dict_to_json_file(final_method_name_insert_dict,json_file_path)
```
