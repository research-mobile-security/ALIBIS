```python
import os
import pandas as pd
```

```python
def get_subdirectories(path):
    subdirectories = []
    # Iterate through all entries in the directory
    for entry in os.listdir(path):
        # Construct the full path of the entry
        full_path = os.path.join(path, entry)
        # Check if the entry is a directory
        if os.path.isdir(full_path):
            # Add the directory to the list of subdirectories
            subdirectories.append(full_path)
    return subdirectories
def get_apk_name(directory_path):
    # Get the directory name from the path
    directory_name = os.path.basename(directory_path)
    # Replace "-java" with ".apk"
    apk_name = directory_name.replace("-java", ".apk")
    return apk_name
def filter_csv_by_apk_names(input_csv_path, output_csv_path, apk_names):
    # Read the CSV file into a DataFrame
    df = pd.read_csv(input_csv_path)
    
    # Filter rows based on the 'apkfile_name' column
    filtered_df = df[df['apkfile_name'].isin(apk_names)]
    
    # Export the filtered DataFrame to a new CSV file
    filtered_df.to_csv(output_csv_path, index=False)
```

```python
# Example usage:
path = r"C:\Users\ASUS\anaconda3\metaLeak-ml-code-search\decompile-code-no-leak"
directories_list = get_subdirectories(path)
#print(directories_list)
apk_file_name_arr = []
for i in range(len(directories_list)):
    apk_file_name = get_apk_name(directories_list[i])
    apk_file_name_arr.append(apk_file_name)
print(apk_file_name_arr)
print(len(apk_file_name_arr))
```

```python
filter_csv_by_apk_names("no_leak_apk.csv", "no_leak_apk_low_size.csv", apk_file_name_arr)
```
