```python
import os
import os
import shutil
import pandas as pd
```

```python
def get_filenames_with_non_ascii(directory):
    non_ascii_filenames = []
    remainder_filenames = []

    # Iterate over all files in the directory
    for filename in os.listdir(directory):
        if not filename.isascii():
            non_ascii_filenames.append(filename)
        else:
            remainder_filenames.append(filename)

    return non_ascii_filenames, remainder_filenames
def get_lowest_size_files(directory, num_files=1200):
    # Get list of all files in directory
    all_files = [(f, os.path.getsize(os.path.join(directory, f))) for f in os.listdir(directory) if os.path.isfile(os.path.join(directory, f))]

    # Sort files by size
    sorted_files = sorted(all_files, key=lambda x: x[1])

    # Get the first num_files files with the lowest size
    lowest_size_files = sorted_files[:num_files]

    # Return only the filenames
    return [file[0] for file in lowest_size_files]
def copy_apk_file(apk_file, source_dir, dest_dir):

    # Check if the APK file exists in the source directory
    src_path = os.path.join(source_dir, apk_file)
    dest_path = os.path.join(dest_dir, apk_file)

    if os.path.exists(src_path):
        # Check if the file exists in the destination directory
        if not os.path.exists(dest_path):
            # Copy the APK file to the destination directory
            shutil.copyfile(src_path, dest_path)
            print(f"Copied {apk_file} to {dest_dir}")
        else:
            print(f"{apk_file} already exists in {dest_dir}")
    else:
        print(f"{apk_file} does not exist in {source_dir}")
def filter_csv_by_apk_names(input_csv, output_csv, apk_names):
    # Read the input CSV file
    df = pd.read_csv(input_csv)

    # Filter rows with apkfile_name in the list lowest_size_files
    filtered_df = df[df['apkfile_name'].isin(apk_names)]

    # Write the filtered DataFrame to a new CSV file
    filtered_df.to_csv(output_csv, index=False)
```
